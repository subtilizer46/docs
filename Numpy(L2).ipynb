```python
import matplotlib.pyplot as plt
import numpy as np

expenses = np.array([300,200,60,10000])
cat = ["Fast Food","Extra","Travelling","Saving"]
colors = ["Red","Blue","Green","Lightblue"]

plt.pie(expenses,labels=cat, colors=colors)
plt.show()
```
