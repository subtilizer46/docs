```python
import pandas

data={
    'Name':['Ram','Alex','Krish','Salmaan'],
    'Age':[20,30,50,90],
    'City':['Solapur','Guntur','Barely','Nagpur']
}
df4=pandas.DataFrame(data)
```

```python
def ca(age):
    if age<=25:
        return 'Young'
    elif age <=40:
        return 'Middle-aged'
    else:
        return 'Senior'
    
df4['Age_Group']=df4['Age'].apply(ca)
```
