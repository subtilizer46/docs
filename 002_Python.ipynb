```python
def greet(name):
    return name+" sharma"

res=greet("Alex")
print(res)
```

```python
def decor(fun):
    def wrapper(name):
        result = fun(name)
        return f"Hello, {result}!"
    return wrapper


@decor
def greet(name):
    return name+" Sharma"

res=greet("Alex")
#print(res)
```

```python
def decor(fun):
    def wrapper(a):
        result = fun(a)
        return result+10
    return wrapper


@decor
def greet(a):
    return a+20

res=greet(30)
print(res)
```

```python
def authenticate(fun):
    def wrapper(username,password):
        if username=="admin" and password=="password":
            return fun(username)
        else:
            return "Authentication Failed"
    return wrapper

@authenticate
def users(username):
    return f"Hello,{username}, Access Granted"

user=input("Enter your username: ")
passw=input("Enter the password: ")

result=users(user,passw)
print(result)
```
